```python
import pandas as pd
import numpy as np
from dataprep.clean import clean_ip
from datetime import datetime
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.feature_selection import VarianceThreshold

import ipaddress

from google.colab import drive
```

```python
drive.mount('/content/drive/')
unprocessed = pd.read_csv('/content/drive/My Drive/DrDoS_UDP.csv')
j=0
unprocessed = unprocessed[j:80000]
cleaned = False
# columns = unprocessed.columns
```

```python
unprocessed.shape
```

```python
unprocessed.rename(columns=lambda x: x.lstrip(), inplace=True)
```

```python
unprocessed.replace([np.inf, -np.inf], 0, inplace=True)

if not cleaned:
  unprocessed = unprocessed.drop(['Flow ID'], axis=1)
  unprocessed = unprocessed.drop(['SimillarHTTP'], axis=1)

  for i in range(j,j+len(unprocessed['Source IP'])):
    if i%10000==0:
      print(str((i-j)/len(unprocessed['Label'])))
    if unprocessed['Label'][i]=="BENIGN": #Convert label to int value 1=DDoS, 0=Benign
      unprocessed['Label'][i]=0
    else:
      unprocessed['Label'][i]=1
    unprocessed['Timestamp'][i] = int(datetime.fromisoformat(unprocessed['Timestamp'][i]).timestamp())
    unprocessed['Source IP'][i] = int(ipaddress.ip_address(unprocessed['Source IP'][i]))
    unprocessed['Destination IP'][i] = int(ipaddress.ip_address(unprocessed['Destination IP'][i]))
clean = True
unprocessed = unprocessed.fillna(0)
unprocessed.head()

```

```python
sel = VarianceThreshold(threshold=(.7 * (1 - .7)))
unprocessed_subset = unprocessed.iloc[:, :-1].copy()
unprocessed_subset.head()
# unprocessed_subset.describe()
#  = sel.fit_transform(unprocessed_subset)
# unprocessed_subset = pd.DataFrame.from_records(unprocessed_subset)
```

```python

unprocessed_subset = sel.fit_transform(unprocessed_subset)
unprocessed_subset = pd.DataFrame.from_records(unprocessed_subset)
```

```python
unprocessed_label = unprocessed['Label']
unprocessed = pd.concat([unprocessed_subset, unprocessed_label], axis=1)
unprocessed.head()
```
