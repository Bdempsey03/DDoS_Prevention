```python
import pandas as pd
col_names = [
'Inbound','ACK Flag Count', 'Timestamp', 'Destination IP', 'Source Port', 'Min Packet Length','Destination IP', 'Protocol',
'Avg Fwd Segment Size', 'Fwd Packet Length Min', 'Packet Length Min', 'Fwd Packet Length mean', 'Fwd Packet Length Max', 'Max Packet Length'
, 'Average Packet Size', 'CWE Flag Count']
UDP = pd.read_csv("DrDoS_UDP.csv", header = 0, names = col_names, dtype = 'unicode')

```

```python
num_of_rows = len(UDP['Destination IP'])

# for x in range (0, num_of_rows):
#     UDP.loc[x, 'Source IP'] = UDP['Source IP'][x].replace(".","")
#     if x%1000000 == 0:
#         print( str((round(x/len(UDP['Source IP']),3)*100)) +"% completed Source IP")
```

```python
for x in range (0, num_of_rows):
    UDP.loc[x, 'Destination IP'] = UDP['Destination IP'][x].replace(".","")
    if x%1000000 == 0:
        print( str((round(x/len(UDP['Destination IP']),3)*100)) +"% completed Destination IP")
```

```python
map = {'BENIGN': 0, 'DrDoS_UDP': 1}
UDP.replace(map, inplace=True)
# for x in range(0, num_of_rows):
#     if x%100000 == 0:
#         print( str((round(x/len(UDP['Destination IP']),3)*100)) +"% completed Label")
    
print(UDP['Timestamp'][1])
UDP['Timestamp'] = pd.to_datetime(UDP['Timestamp']).astype('int64')/10**9
```

```python
UDP
```

```python
#drop all zero columns
UDP = UDP.replace('Infinity', None)
UDP = UDP.dropna()
UDP = UDP.loc[(UDP!=0).any(axis=1)]
UDP['Source IP'], UDP['Destination IP'] = UDP['Source IP'].astype('int64'), UDP['Destination IP'].astype('int64')
UDP[col_names] = UDP[col_names].astype('float64')
UDP.head()
```

```python
UDP = UDP['Inbound','ACK Flag Count', 'Timestamp', 'Destination IP', 'Source Port', 'Min Packet Length','Destination IP', 'Protocol',
'Avg Fwd Segment Size', 'Fwd Packet Length Min', 'Packet Length Min', 'Fwd Packet Length mean', 'Fwd Packet Length Max', 'Max Packet Length'
, 'Average Packet Size', 'CWE Flag Count']
```

```python
UDP.to_csv("cleaned_DrDoS_UDP.csv")
```
